```python
from torch.utils.data import Dataset    #导入工具
'''PIL / Pillow 是 Python 里最常用的 图像处理库
Image是它最核心的模块，负责“打开、创建、修改、保存图片'''
from PIL import Image
import os

class MyData(Dataset):
    def __init__(self,root_dir,lable_dir):
        self.root_dir = root_dir
        self.lable_dir = lable_dir
        self.path = os.path.join(self.root_dir,self.lable_dir)
        self.img_path = os.listdir(self.path)   #获得所有图片的地址

    def __getitem__(self, idx):
        img_name=self.img_path[idx]     #得到某一张图片的地址
        img_item_path=os.path.join(self.root_dir,self.lable_dir,img_name)   #拼接出图片的完整地址
        img=Image.open(img_item_path)
        lable=self.lable_dir
        return img,lable

    def __len__(self):
        return len(self.img_path)   #长度就是图片地址列表的长度，也就是有多少张图片
```

```python
#注：实验的数据集以文件名称直接作为标签
root_dir="E:\\DataSet\\ants_bee_dataset\\train"
ants_lable_dir="ants"
bees_lable_dir="bees"
ants_dataset=MyData(root_dir,ants_lable_dir)
bees_dataset=MyData(root_dir,bees_lable_dir)
train_dataset=Dataset.__add__(ants_dataset,bees_dataset)    #也可直接用'+'号
img,lable=train_dataset.__getitem__(200)
img.show()
print(len(train_dataset))
```
